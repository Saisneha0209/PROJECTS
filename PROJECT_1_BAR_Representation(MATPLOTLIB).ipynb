#SALES FOR DIFFERENT STATES AND THEIR PRODUCT SALES WITH BAR GRAPH(TO REPRESENT THEM)

```python
import matplotlib.pyplot as plt
customer_name=["Damini","Kiara","Vineet","Preeti","Shree","Sudeeksha","Shyam","Krishna"]
states=["Odisha","Chandigarh","Odisha","Gujarat","Odisha","Shimla","Bihar","Bihar"]
product=["shirt","pant","socks","hat","sunglass","scrunchies","crocks", "shirt"]
amount=[450,520,120,455,630,250,450,140]
gender=["male","female"]
total_sales=0
for i in amount: #1st analysis for total amount(sales)
  total_sales=total_sales+i
print(total_sales)
unique_state=[]
sales_state=[]
sales_product=[]
profit_state=[]
profit_product=[]
for i in states: #2nd analysis for the states(We do not want states to come 2 times and we can not count 2 times so for this)
  if i not in unique_state:
   unique_state.append(i)
   sales_state.append(0)
print(unique_state) # Print unique states

#product wise
unique_product=[]
for i in product:
  if i not in unique_product:
    unique_product.append(i)
    sales_product.append(0)
print(unique_product)

#gender wise
unique_gender=[]
for i in gender:
  if i not in unique_gender:
    unique_gender.append(i)
print(unique_gender)

 #state wise sales calculation
for i in range(len(states)):
     index=unique_state.index(states[i])
     sales_state[index]=sales_state[index]+amount[i]
print(sales_state)
 #product wise sales calculation
for i in range (len(product)):
  index=unique_product.index(product[i])
  sales_product[index]=sales_product[index]+amount[i]
print(sales_product)

#profit-wise sales calculation
for i in range (len(states)):
  index=unique_state.index(states[i])
  profit_state[index]=sales_state[index]+amount[i]
print(profit_state)


plt.bar(unique_state, sales_state)
plt.title("Representation of sales in states")
plt.xlabel("States")
plt.ylabel("Sales Amount")
plt.show()


plt.bar(unique_product,sales_product)
plt.title("Representation of sales by product")
plt.xlabel("Products")
plt.ylabel("Sales Amount")
```

```python
# Assuming a hypothetical 30% profit margin for demonstration
profit_margin = 0.30

# Calculate profit for each sale
profit_amount = [amount[i] * profit_margin for i in range(len(amount))]

# State-wise profit calculation
profit_state = [0] * len(unique_state)
for i in range(len(states)):
    index = unique_state.index(states[i])
    profit_state[index] = profit_state[index] + profit_amount[i]

print("State-wise Profit:")
print(profit_state)

# Product-wise profit calculation
profit_product = [0] * len(unique_product)
for i in range(len(product)):
    index = unique_product.index(product[i])
    profit_product[index] = profit_product[index] + profit_amount[i]

print("Product-wise Profit:")
print(profit_product)

# Plotting profit by state
plt.bar(unique_state, profit_state)
plt.title("Representation of Profit by States")
plt.xlabel("States")
plt.ylabel("Profit Amount")
plt.show()

# Plotting profit by product
plt.bar(unique_product, profit_product)
plt.title("Representation of Profit by Product")
plt.xlabel("Products")
plt.ylabel("Profit Amount")
plt.show()
```
