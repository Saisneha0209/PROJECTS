#Mobile Recharge Reminder:-


```python
balance = int(input("Enter your current balance: "))
while balance >= 50:
    balance = balance - 50
    print("Remaining balance:", balance)
print("Recharge needed")
```

#Checking the availability of SEATS!

```python
# Total seats fixed
total_seats = 50
tickets = int(input("How many tickets do you want? "))
# Check availability
if tickets <= total_seats:
    total_seats = total_seats - tickets
    print("Tickets booked successfully!")
    print("Remaining seats:", total_seats)
else:
    print("Sorry! No more seats are actually available!!")
```

#Calculating full price of prooducts!

```python
cart_items = [
    ("Laptop", 55000, 1),
    ("Headphones", 2000, 2),
    ("Mouse", 800, 1),
    ("Keyboard", 1500, 1)
]

def calculate_item_total(price, quantity):
    total_price = price * quantity
    return total_price

grand_total = 0
for item in cart_items:
    name, price, quantity = item
    item_total = calculate_item_total(price, quantity)
    print(name, "Total:", item_total)
    grand_total = grand_total + item_total

print("Grand Total:", grand_total)
```

#HOTEL BILL:-

```python
orders = [
    ("Burger", 120, 2),
    ("Pizza", 350, 1),
    ("Cold Drink", 50, 3),
    ("Fries", 90, 1)
]

distance = 5   # delivery distance in km


#here we will calculate_item_total
def calculate_item_total(price, quantity):
    return price * quantity


#here we will generate_order_summary
def generate_order_summary(orders):
    summary = []
    for item in orders:
        name, price, quantity = item
        total_price = calculate_item_total(price, quantity)
        summary.append((name, total_price))
    return summary


#here we will calculate_food_total
def calculate_food_total(order_summary):
    total = 0
    for item in order_summary:
        total += item[1]
    return total


#here we will calculate_delivery_charge
def calculate_delivery_charge(distance):
    if distance <= 3:
        return 30
    elif distance <= 6:
        return 50
    else:
        return 80

def apply_discount(food_total):
    if food_total >= 500:
        return food_total * 0.15
    else:
        return 0

order_summary = generate_order_summary(orders)
food_total = calculate_food_total(order_summary)
discount = apply_discount(food_total)
delivery_charge = calculate_delivery_charge(distance)
final_amount = food_total - discount + delivery_charge

print("------ ORDER SUMMARY ------")
for item in order_summary:
    print(item[0], ": ‚Çπ", item[1])

print("---------------------------")
print("Food Total : ‚Çπ", food_total)
print("Discount (15%) : ‚Çπ", int(discount))
print("Delivery Charge : ‚Çπ", delivery_charge)
print("---------------------------")
print("Final Amount Payable : ‚Çπ", int(final_amount))
print("---------------------------")
print("Enjoy your meal üçΩÔ∏è")
```
